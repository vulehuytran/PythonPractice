```python
#input data 3

grades = [83, 95, 91, 87, 70, 0, 85, 82, 100, 67, 73, 77, 0]
```

```python
#want to plot histogram. Bar graph has 2 variables, but histogram has only 1 variable.

#step 1: divide the entire range of values into a series of intervals. one intervial=10

def decile(grade):
    return grade//10*10

#step 2: count how many values fall into each interval

from collections import Counter
histogram = Counter(decile(grade) for grade in grades)
```

```python
#histogram is now a dictionary with keys and values
print(histogram)
print(histogram.keys())
print(histogram.values())
```

```python
plt.bar(histogram.keys(), histogram.values())
plt.title("Distribution of Exam 1 Grades")
plt.xlabel("Decile")
plt.ylabel("# of Students")
plt.show()
```

```python
#now I will need to make the histogram more beautiful

plt.bar(histogram.keys(), histogram.values(), width=8)
plt.title("Distribution of Exam 1 Grades")
plt.axis([-5, 105, 0, 5])
plt.xticks([10*i for i in range(11)])
plt.xlabel("Decile")
plt.ylabel("# of Students")
plt.show()
```

```python
#LINE CHARTS
#input data 5

variance = [1, 2, 4, 8, 16, 32, 64, 128, 256]
bias_squared = [256, 128, 64, 32, 16, 8, 4, 2, 1]
```

```python
total_error = [x + y for x, y in zip(variance, bias_squared)]
```

```python
#now I need something for my x axis

xs = [i for i in range(len(variance))]
print(xs)
```

```python
plt.plot(xs, variance)
plt.plot(xs, bias_squared)
plt.plot(xs, total_error)
```

```python
plt.plot(xs, variance)
plt.plot(xs, bias_squared)
plt.plot(xs, total_error)

plt.legend()
plt.title("The bias-variance tradeoff")
plt.xlabel("model complexity")

#by default, legend() will give a legend on the upper right location.
#because I didn't label each line, there is no info in the legend
```

```python
plt.plot(xs, variance, label='variance')
plt.plot(xs, bias_squared, label='bias^2')
plt.plot(xs, total_error, label='total error')

plt.legend(loc = 9)
plt.title("The bias-variance tradeoff")
plt.xlabel("model complexity")
```

```python
plt.plot(xs, variance, 'g-', label='variance')
plt.plot(xs, bias_squared, 'r-.', label='bias^2')
plt.plot(xs, total_error, 'b:', label='total error')

plt.legend(loc = 9)
plt.title("The bias-variance tradeoff")
plt.xlabel("model complexity")
```

```python
# SCATTERPLOTS

#input data 6

friends = [70, 65, 72, 63, 71, 64, 60, 64, 67]
minutes = [175, 170, 205, 120, 220, 130, 105, 145, 190]
```

```python
plt.scatter(friends, minutes)

plt.title("Daily Minutes vs Number of Friends")
plt.xlabel("# of friends")
plt.ylabel("daily minutes spent on the site")
plt.show()
```

```python
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i']
```

```python
# label each data point

for label, friend_count, minute_count in zip(labels, friends, minutes):
    plt.annotate(label, xy=(friend_count, minute_count), xytext=(5, -5), textcoords='offset points')

plt.scatter(friends, minutes)
plt.axis([0, 100, 0, 230])
plt.title("Daily Minutes vs Number of Friends")
plt.xlabel("# of friends")
plt.ylabel("daily minutes spent on the site")
plt.show()
```
